```python
# Importar as libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
#Ler os dados
Base_Dados = pd.read_csv('/content/unicorns till sep 2022.csv')

# Verificar Dimensão
Base_Dados.shape
```

```python
# Primeiros Registros
Base_Dados.head()
```

```python
# Colunas
Base_Dados.columns
```

```python
# Renomear
Base_Dados.rename( columns={
    'Unnamed: 0':'Id',
    'Company':'Empresa',
    'Valuation ($B)': 'Valor ($)',
    'Date Joined':'Data de Adesão',
    'Country':'Pais',
    'City':'Cidade',
    'Industry':'Setor',
    'Investors':'Investidores',
}, inplace=True

)
```

```python
# Verificar Tipo da Informação
Base_Dados.info()
```

```python
# Campos nulos
Base_Dados.isnull().sum()
```

```python
# Grafica
plt.figure( figsize=(15,6))
plt.title('Analisando Campos Nulos')
sns.heatmap( Base_Dados.isnull(), cbar=False);
```

```python
# Campos unicos
Base_Dados.nunique()
```

```python
# Valores Unicos
Base_Dados['Setor'].unique()
```

```python
# Valores Unicos - Rank
Base_Dados['Setor'].value_counts()
```

```python
# Valores Unicos - Rank
Base_Dados['Setor'].value_counts( normalize=True)
```

```python
plt.figure( figsize=(15,6))
plt.title('Analise dos Setores')
plt.bar( Base_Dados['Setor'].value_counts().index, Base_Dados["Setor"].value_counts())
plt.xticks( rotation=45, ha='right');
```

```python
# Plot geral dos Paises
import seaborn as sns

Analise = round(Base_Dados['Pais'].value_counts(normalize=True) * 100, 1).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(
    x=Analise.values,
    y=Analise.index,
    palette='viridis'
)
plt.title('Análise dos Países Geradores de Unicórnios - Top 10')
plt.xlabel('Porcentagem (%)')
plt.ylabel('Países')
plt.show()
```

```python
# conversão para Data
Base_Dados['Data de Adesão'] = pd.to_datetime( Base_Dados['Data de Adesão'])

Base_Dados['Data de Adesão'].head()
```

```python
# Extrair o Ano e Mês
Base_Dados['Mês'] = pd.DatetimeIndex( Base_Dados['Data de Adesão']).month
Base_Dados['Ano'] = pd.DatetimeIndex( Base_Dados['Data de Adesão']).year

Base_Dados.head()
```

```python
# Tabela Analitica
Analise_Agrupada = Base_Dados.groupby(['Pais', 'Ano', 'Mês', 'Empresa']).size().reset_index(name='Total')

Analise_Agrupada
```

```python
Analise_Agrupada.loc[Analise_Agrupada['Pais'] == 'Brazil']
```

```python
# Transformando a coluna valor
Base_Dados['Valor ($)'] = pd.to_numeric(
    Base_Dados['Valor ($)'].astype(str).apply(lambda x: x.replace('$', '').replace(',', '')),
    errors='coerce'
)

Base_Dados[['Valor ($)']].head()

```

```python
# Tabela Analitica
Analise_Pais = (
    Base_Dados
    .groupby('Pais')['Valor ($)']
    .sum()
    .reset_index()
)
Analise_Pais.head()
```

```python
Analise_Valor = Analise_Pais.sort_values('Valor ($)', ascending=False)
Analise_Valor.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot( Analise_Valor['Pais'], Analise_Valor['Valor ($)'])
plt.title('Analise do Valor por Pais')
plt.xticks( rotation=45, ha='right');
```
